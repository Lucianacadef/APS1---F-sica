APS1 - Pêndulo


```python
import numpy as np
from scipy.integrate import odeint
import matplotlib.pyplot as plt
import math as m
from scipy.constants import *
%matplotlib inline
```

Ângulo de 30 graus

```python
#lista tempo
dt = 0.001
tempo = np.arange(0,18,dt)

#EqDif do pêndulo

def EqDif(Sol, t):
    ang_t = Sol[0]
    ang_o = Sol[1]    
    dang_tdt = ang_o
    dang_odt = -(g*m.sin(ang_t))/L
    return [dang_tdt, dang_odt]

#ângulo inicial e comprimento do fio
ang = 30*m.pi/180
S0=[ang,0]
L = 2 
```

```python
Sol=odeint(EqDif, S0, tempo)

X = []
Y = []
for i in Sol[:,0]:
    x = 2*m.sin(i)
    y = 2 - 2*m.cos(i)
    X.append(x)
    Y.append(y)
    
#Plotagem

plt.plot(X, Y)
plt.title('eixo X por eixo Y')
plt.grid(True)
plt.show()

plt.plot(tempo, X)
plt.title('eixo X pelo tempo')
plt.grid(True)
plt.show()

plt.plot(tempo, Y)
plt.title('eixo Y pelo tempo')
plt.grid(True)
plt.show()

plt.plot(tempo, Sol[:,0]*180/m.pi)
plt.title('ângulo pelo tempo')
plt.grid(True)
plt.show()
```
